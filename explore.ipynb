```python
from text_processing.text_processing import (load_docs,remove_stopwords, normalise_text, lemmatize)
```

```python
doc_files = [
    "example_docs_for_backend_test/test docs/doc1.txt",
    "example_docs_for_backend_test/test docs/doc2.txt",
    "example_docs_for_backend_test/test docs/doc3.txt",
    "example_docs_for_backend_test/test docs/doc4.txt",
    "example_docs_for_backend_test/test docs/doc5.txt",
    "example_docs_for_backend_test/test docs/doc6.txt",
]
docs = load_docs(doc_files)
docs[0].sentences[:5]
```

```python
for doc in docs:
    for idx, sentence in enumerate(doc.sentences):
        tokens = normalise_text(sentence)
        tokens = remove_stopwords(tokens)
        tokens = lemmatize(tokens)
        doc.processed_sentences.append(tokens)
        print(idx, " " ,tokens)
```
